Change the body part and location parts to get a "mask" for when the conditions specified are met. 

First, make a mask to ignore any frames where the position of the bodypart is uncertain. Choose a liklihood threshold (all values above it are counted).

```python
probability_threshold = 0.6

mask_likelihood = df['body_center']['likelihood']>probability_threshold
```


Calculate the fraction of time the specified condition was met. 

Print a statement including the result of the calculation.

```python
mask = (df['body_center']['y']>df['cbl']['y']) | (df['body_center']['y']<df['ctl']['y'])
mask = mask & mask_likelihood
frac_time_in = sum(mask)/len(df)
print(f'the fraction of time spent in the open arm is {frac_time_in:0.2f}')

# this mask should include time spent within center closed area
# mask = (df['body_center']['y']<df['cbl']['y']) & (df['body_center']['y']>df['ctl']['y'])

# this mask should only be when the animal is observed in the closed parts of the arm explicitly
mask = (df['body_center']['x']<df['cbl']['x']) | (df['body_center']['x']>df['cbr']['x'])
mask = mask & mask_likelihood
frac_time_in = sum(mask)/len(df)
print(f'the fraction of time spent in the closed arm is {frac_time_in:0.2f}')

# NOTE: given how you have labelled, the second two masks I have commented should come up with the same answer,
# plotting the locations included in each masked condition will help understand if something is incorrect.
```
